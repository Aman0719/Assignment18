```python
Q3. What is App routing in Flask? Why do we use app routes?

Ans-

In Flask, app routing refers to the process of mapping a URL endpoint to a Python function that will handle the request to that endpoint. 
The @app.route decorator is used to specify which URL endpoint should be handled by which function.

For example, consider the following Flask app route:
    
@app.route('/hello')
def hello():
    return 'Hello, World!'


In this example, the @app.route('/hello') decorator maps the URL endpoint /hello to the hello() function. 
When a user navigates to the URL http://example.com/hello, Flask will call the hello() function and return the string "Hello, World!" to the user's browser.

We use app routes in Flask to create the different pages and functionalities of our web application.
By mapping each URL endpoint to a specific function, we can create a logical and organized structure for our web app. 
It also makes it easy to update and modify the functionality of specific endpoints as needed.

In addition, app routing in Flask allows us to create RESTful APIs, which provide a way for other applications to interact with our web app and exchange data.
We can use app routes to define the endpoints for different API calls and specify how the data should be returned.
```

```python
Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999

Ans-

To get our output, we need to write code in the flask lab and run it there:
    
The code for the above is:
    
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route("/")
def company_details():
    return "<h1>Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999</h1>"
    

if __name__=="__main__":
    app.run(host="0.0.0.0")
    
In this example, the /welcome route is mapped to the welcome() function, which simply returns the welcome message "Welcome to ABC Corporation". 
The / route is mapped to the company_details() function, which returns an HTML page with the company details.

To navigate to welcome function:
https://red-photographer-cfyab.pwskills.app:5000/welcome

To navigate to the / function:-
https://red-photographer-cfyab.pwskills.app:5000/

Here are screen shots of both
```

```python
Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

Ans-

In Flask, the url_for() function is used for URL building. 
It is a helper function that generates a URL to a specific function or endpoint based on the name of the function and the parameters passed to it.

Here is an example of how to use the url_for() function in Flask:

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return '''
    <h1>Welcome to my Flask App</h1>
    <p>Click <a href="{}">here</a> to go to the about page</p>
    '''.format(url_for('about'))

@app.route('/about')
def about():
    return 'This is the about page'

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

```python
In this example, the index() function returns an HTML page with a link to the /about route.
Instead of hardcoding the URL for the link, we use the url_for() function to generate the URL dynamically based on the name of the about() function.

When you run this application and navigate to app route /, you should see the index page with a link to the about page. 
When you click on the link, you should be taken to the /about route and see the message "This is the about page".
```
